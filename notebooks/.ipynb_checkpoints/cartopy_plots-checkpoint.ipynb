```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
import numpy as np
```

```python
figsize = (10,20)
dpi     = 100
LL, UR  = (-4.0, -34.0), (6.0, -28.0)
color   = 'red'

plt.figure(figsize=figsize, dpi=dpi)
ax = plt.axes(projection=ccrs.NearsidePerspective(central_longitude=15.0, 
                                                  central_latitude=-8.0,
                                                  satellite_height=4.0e7))
ax.coastlines()
ax.gridlines()
ax.add_feature(cfeature.LAND, color='lightgray')


ax.set_global()

plt.plot([LL[0], LL[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )

plt.plot([UR[0], UR[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )

plt.plot([LL[0], UR[0]], [LL[1], LL[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )

plt.plot([LL[0], UR[0]], [UR[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),         )


plt.show()
```

```python
figsize = (10,20)
dpi     = 100
LL, UR  = (-4.0, -34.0), (6.0, -28.0)
color   = 'black'

grid_x = 15
grid_y = 15



perspective = ccrs.NearsidePerspective(central_longitude=15.0, 
                                       central_latitude=-8.0,
                                       satellite_height=4.0e7)

plt.figure(figsize=figsize, dpi=dpi)
ax = plt.axes(projection=perspective)
ax.coastlines()
ax.gridlines()


ax.add_feature(cfeature.LAND, color='lightgray')
ax.set_extent([-40, 20, -50, -20], crs=ccrs.PlateCarree())


# plot bounding boxes
plt.plot([LL[0], LL[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([UR[0], UR[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [LL[1], LL[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [UR[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),         )

# plot starting mesh
lon = np.linspace(LL[0],UR[0],grid_x)
lat = np.linspace(LL[1],UR[1],grid_y)
for i in range(grid_x-1):
    for j in range(grid_y-1):
        plt.plot([lon[i], lon[i+1]], [lat[j], lat[j]],
        color=color, linewidth=1, marker=None,
        transform=ccrs.PlateCarree(),         )
        
        plt.plot([lon[i], lon[i]], [lat[j], lat[j+1]],
        color=color, linewidth=1, marker=None,
        transform=ccrs.PlateCarree(),         )
        
        if j>0:
            plt.plot([lon[i], lon[i+1]], [lat[j], lat[j-1]],
            color=color, linewidth=1, marker=None,
            transform=ccrs.PlateCarree(),         )
          
    plt.plot([lon[i], lon[i+1]], [lat[grid_x-1], lat[grid_x-2]],
             color=color, linewidth=1, marker=None,
             transform=ccrs.PlateCarree(),         )

plt.show()
```

```python
import csv
import itertools
```

```python
path_prefix = "D:\\Documents\\Uni\\3 Mathematics in Data Science\\05 Masterarbeit\\Bilder"
nodes = []
cells = []
with open(path_prefix+"\\nodes.csv", newline='') as nodes_file:
    nodes_reader = csv.reader(nodes_file, delimiter=',', quotechar='|')
    nodes = [list(map(float, row)) for row in nodes_reader]
with open(path_prefix+"\\cells.csv", newline='') as cells_file:
    cells_reader = csv.reader(cells_file, delimiter=',', quotechar='|')
    cells = [list(map(int, row)) for row in cells_reader]
```

```python
figsize    = (10,20)
dpi        = 100
LL, UR     = (-4.0, -34.0), (6.0, -28.0)
color      = 'lightgray'
color_grid = 'black'

grid_x = 15
grid_y = 15



perspective = ccrs.NearsidePerspective(central_longitude=15.0, 
                                       central_latitude=-8.0,
                                       satellite_height=4.0e7)

plt.figure(figsize=figsize, dpi=dpi)
ax = plt.axes(projection=perspective)
ax.coastlines()
ax.gridlines()


ax.add_feature(cfeature.LAND, color='lightgray')
ax.set_extent([-40, 20, -50, -20], crs=ccrs.PlateCarree())


# plot bounding boxes
plt.plot([LL[0], LL[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([UR[0], UR[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [LL[1], LL[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [UR[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),         )

    
# plot the new mesh    
for cell in cells:
    for node_num_a, node_num_b in itertools.combinations(cell, 2):
        node_a = nodes[node_num_a-1]
        node_b = nodes[node_num_b-1]
        plt.plot([node_a[0],node_b[0]], [node_a[1],node_b[1]],
        color=color_grid, linewidth=1, marker=None,
        transform=ccrs.PlateCarree(),         )

plt.show()
```

```python
path_prefix = "D:\\Documents\\Uni\\3 Mathematics in Data Science\\05 Masterarbeit\\Bilder"
initial = []
final = []
wind = []
with open(path_prefix+"\\733010.0.csv", newline='') as initial_file:
    initial_reader = csv.reader(initial_file, delimiter=',', quotechar='|')
    initial = [list(map(float, row)) for row in initial_reader]
with open(path_prefix+"\\733090.0.csv", newline='') as final_file:
    final_reader = csv.reader(final_file, delimiter=',', quotechar='|')
    final = [list(map(float, row)) for row in final_reader]
with open(path_prefix+"\\wind.csv", newline='') as wind_file:
    wind_reader = csv.reader(wind_file, delimiter=',', quotechar='|')
    wind = [list(map(float, row)) for row in wind_reader]
```

```python
Lon = np.arange(start=-12.0,stop=29.25,step=0.25)
Lat = np.arange(start=-45.0,stop=-18.75,step=0.25)
initial_lon = np.asarray([row[::2] for row in initial])
initial_lat = np.asarray([row[1::2] for row in initial])
final_lon = np.asarray([row[::2] for row in final])
final_lat = np.asarray([row[1::2] for row in final])
wind_lon = np.asarray([row[::2] for row in wind])
wind_lat = np.asarray([row[1::2] for row in wind])
```

```python
figsize    = (10,20)
dpi        = 100
LL, UR     = (-4.0, -34.0), (6.0, -28.0)
color      = 'lightgray'
color_arrows = 'black'
scale      = 10

perspective = ccrs.PlateCarree()

plt.figure(figsize=figsize, dpi=dpi)
ax = plt.axes(projection=perspective)
ax.coastlines()
#ax.gridlines()


ax.add_feature(cfeature.LAND, color='lightgray')
ax.set_extent([-10, 10, -35, -25], crs=ccrs.PlateCarree())


# plot bounding boxes
plt.plot([LL[0], LL[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([UR[0], UR[0]], [LL[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [LL[1], LL[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),
         )
plt.plot([LL[0], UR[0]], [UR[1], UR[1]],
         color=color, linewidth=1, marker=None,
         transform=ccrs.PlateCarree(),         )

    
# plot the stream lines
skip=(slice(None,None,2),slice(None,None,2))
ax.quiver(Lon[::2],Lat[::2],initial_lon[skip],initial_lat[skip],color=color_arrows,scale=scale,transform=ccrs.PlateCarree(),)

plt.show()
```
